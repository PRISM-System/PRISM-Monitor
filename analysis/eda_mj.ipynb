```python
# =============================================================================
# 11. 전체 데이터셋 요약 정보
# =============================================================================

print("\n" + "="*60)
print("=== 전체 데이터셋 요약 ===")

datasets = [
    ('SEMI_PARAM_MEASURE', df_param),
    ('SEMI_LOT_MANAGE', df_lot),
    ('SEMI_PHOTO_SENSORS', df_photo),
    ('SEMI_CMP_SENSORS', df_cmp),
    ('SEMI_CVD_SENSORS', df_cvd),
    ('SEMI_ETCH_SENSORS', df_etch),
    ('SEMI_IMPLANT_SENSORS', df_implant),
    ('SEMI_EQUIPMENT_SENSOR', df_equipment),
    ('SEMI_PROCESS_HISTORY', df_process)
]

print(f"{'Dataset Name':<25} {'Rows':<8} {'Columns':<8}")
print("-" * 45)
for name, df in datasets:
    print(f"{name:<25} {len(df):<8} {len(df.columns):<8}")

# =============================================================================
# 12. 결측값 확인
# =============================================================================

print("\n" + "="*60)
print("=== 각 데이터셋별 결측값 확인 ===")

for name, df in datasets:
    missing_count = df.isnull().sum().sum()
    print(f"{name}: 총 {missing_count}개 결측값")
    
    if missing_count > 0:
        missing_by_column = df.isnull().sum()
        missing_columns = missing_by_column[missing_by_column > 0]
        print(f"  결측값이 있는 컬럼:")
        for col, count in missing_columns.items():
            print(f"    {col}: {count}개")
    print()
```

```python
# =============================================================================
# 13. 숫자형 컬럼 확인
# =============================================================================

print("=== 각 데이터셋별 숫자형 컬럼 확인 ===")

for name, df in datasets:
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"\n{name}:")
    print(f"  숫자형 컬럼 개수: {len(numeric_cols)}")
    if len(numeric_cols) > 0:
        print(f"  숫자형 컬럼 목록:")
        for i, col in enumerate(numeric_cols):
            print(f"    {i+1:2d}. {col}")
```

```python
# =============================================================================
# 1. SEMI_PARAM_MEASURE 데이터 상세 분석
# =============================================================================

print("=" * 60)
print("SEMI_PARAM_MEASURE 데이터 상세 분석")
print("목적: 스펙 위반/근접 + 타깃에서 얼마나 벗어났는지 확인")
print("=" * 60)

# 데이터 로드 (이전 코드에서 이미 로드했다고 가정)
# df_param = pd.read_csv('path/to/semi_param_measure.csv')

# 기본 통계 확인
print("기본 통계:")
print(df_param.describe())

# 스펙 관련 컬럼이 있는지 확인
spec_columns = ['MEASURED_VAL', 'TARGET_VAL', 'USL', 'LSL']
available_spec_cols = [col for col in spec_columns if col in df_param.columns]
print(f"\n사용 가능한 스펙 관련 컬럼: {available_spec_cols}")

# 숫자형 컬럼들의 분포 확인
numeric_cols = df_param.select_dtypes(include=[np.number]).columns.tolist()
print(f"\n숫자형 컬럼 ({len(numeric_cols)}개): {numeric_cols}")

# 시각화 1: 주요 파라미터들의 분포
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols[:6]):
    if i < len(axes):
        axes[i].hist(df_param[col].dropna(), bins=30, alpha=0.7, edgecolor='black')
        axes[i].set_title(f'{col} 분포')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('빈도')
        
        # 평균선 추가
        mean_val = df_param[col].mean()
        axes[i].axvline(mean_val, color='red', linestyle='--', label=f'평균: {mean_val:.2f}')
        axes[i].legend()

# 빈 subplot 제거
for i in range(len(numeric_cols), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# 시각화 2: 박스플롯으로 이상치 확인
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols[:6]):
    if i < len(axes):
        box_data = df_param[col].dropna()
        axes[i].boxplot(box_data)
        axes[i].set_title(f'{col} 이상치 확인')
        axes[i].set_ylabel(col)
        
        # 이상치 개수 계산
        Q1 = box_data.quantile(0.25)
        Q3 = box_data.quantile(0.75)
        IQR = Q3 - Q1
        outliers = box_data[(box_data < Q1 - 1.5*IQR) | (box_data > Q3 + 1.5*IQR)]
        axes[i].text(0.5, 0.95, f'이상치: {len(outliers)}개', 
                    transform=axes[i].transAxes, ha='center', va='top')

for i in range(len(numeric_cols), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# 스펙 위반 분석 (해당 컬럼이 있는 경우)
if 'MEASURED_VAL' in df_param.columns and 'USL' in df_param.columns and 'LSL' in df_param.columns:
    print("\n스펙 위반 분석:")
    
    # USL 위반
    usl_violations = df_param[df_param['MEASURED_VAL'] > df_param['USL']]
    print(f"USL 위반 건수: {len(usl_violations)}")
    
    # LSL 위반  
    lsl_violations = df_param[df_param['MEASURED_VAL'] < df_param['LSL']]
    print(f"LSL 위반 건수: {len(lsl_violations)}")
    
    # 타깃 편차
    if 'TARGET_VAL' in df_param.columns:
        target_deviation = abs(df_param['MEASURED_VAL'] - df_param['TARGET_VAL'])
        print(f"평균 타깃 편차: {target_deviation.mean():.3f}")
        
        # 스펙 위반 시각화
        plt.figure(figsize=(12, 6))
        plt.scatter(range(len(df_param)), df_param['MEASURED_VAL'], alpha=0.6, s=30)
        plt.axhline(df_param['USL'].iloc[0], color='red', linestyle='--', label='USL')
        plt.axhline(df_param['LSL'].iloc[0], color='red', linestyle='--', label='LSL')
        plt.axhline(df_param['TARGET_VAL'].iloc[0], color='green', linestyle='-', label='Target')
        plt.title('측정값 vs 스펙 한계')
        plt.xlabel('데이터 포인트')
        plt.ylabel('측정값')
        plt.legend()
        plt.show()
```
